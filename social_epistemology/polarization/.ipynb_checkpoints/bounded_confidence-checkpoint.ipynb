```python
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
class PolarizationSimulation:

    def __init__(self, n, sigma, steps):
        self.n = n
        self.sigma = sigma
        self.steps = steps

        self.history = None


    def run(self):

        opinions, history = self.init_params()
        for step in tqdm(range(self.steps-1)):
            new_opinion_profile = self.update(history[:, -1])
            history = np.concatenate([history, new_opinion_profile], axis=1)

        self.history = history


    def init_params(self):

        opinions = np.random.random(size=self.n)
        opinions.sort()
        history = opinions.reshape(-1, 1)

        return opinions, history


    def update(self, opinions: np.array) -> np.array:

        new_opinion_profile = []
        for opinion in opinions:
            neighbors = np.where((opinions > opinion-self.sigma) & \
                                 (opinions < opinion+self.sigma), 
                                 True, False)
            new_opinion = opinions[neighbors].mean()
            new_opinion_profile.append(new_opinion)

        return np.array(new_opinion_profile).reshape(-1, 1)


    def plot_results(self):
        
        fig, ax = plt.subplots(figsize=(17,8))
        cmap = cm.get_cmap('hsv', self.n)
        x = np.arange(self.steps)

        for agent in range(self.n):
            ax.plot(x, self.history[agent, :], color=cmap(agent))
```

```python
isolation = PolarizationSimulation(
    n     = 1000,
    sigma = .01,
    steps = 25
)

isolation.run()
isolation.plot_results()
```

```python
bimodal = PolarizationSimulation(
    n     = 1000,
    sigma = .15,
    steps = 25
)

bimodal.run()
bimodal.plot_results()
```

```python
consensus = PolarizationSimulation(
    n     = 1000,
    sigma = .25,
    steps = 25
)

consensus.run()
consensus.plot_results()
```
